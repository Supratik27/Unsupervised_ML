**K-Means clustering using scikit learn package**

We will take an example with marks of 22 students for 2 subjects.

The problem is to find groups of excellent and under performing students.

dat.png

Visualizing and understanding the intput data

```python
import matplotlib.pyplot as plt

x = [ 5,  4, 13, 11, 13,  5,  3,  4,  3, 10, 12, 13, 14, 12, 11,  6,  7,  6,  5, 11, 14, 10]
y = [15, 18, 22, 21, 21, 19, 15, 16, 16, 21, 23, 22, 22, 20, 22, 17, 18, 14, 17, 24, 22, 23]

plt.scatter(x,y, color="blue")
plt.show()
```

Data processing

```python
data = list(zip(x,y))
print(data)
```

Now we will add the K-Means algorithm with scikit learn library

```python
from sklearn.cluster import KMeans
Kmeans = KMeans(n_clusters=2, n_init='auto') #We will take the 2 clusters as there are 2 subjects
Kmeans.fit(data)

c=Kmeans.labels_
print(c)
```

Visualizing the clusters

```python
plt.scatter(x, y, c=Kmeans.labels_)
plt.show()
```

Clustering evaluation without ground truth

```python
from sklearn import metrics

metrics.silhouette_score(data, c, metric='euclidean')
```

As the value of the "Silhouette Coefficient" is near to +1 therefore, the clusters are highly dense and well separated

```python
metrics.calinski_harabasz_score(data, c)
```

As the value of the "Calinski-Harabasz Index" is higher therefore, the clusters are highly dense and well separated

```python
metrics.davies_bouldin_score(data, c)
```

As the value of the "Davies-Bouldin Index" is nearing to 0 therefore, the clusters have better partition

**We will try with data where it is difficult to determine K (the number of clusters)**

```python
import matplotlib.pyplot as plt

x = [ 4,  5, 10,  4,  3, 11, 14 , 6, 10, 12,  7, 10]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21, 20, 22]

plt.scatter(x, y, color='black')
plt.show()
```

The ELBOW method

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

From the above graph we can observe that the value of K or n_clusters can be 2, 3 or 4 such that we can get a good cluster

Now we will add the K-Means algorithm with scikit learn library and visualize the clusters

```python
from sklearn.cluster import KMeans
data = list(zip(x, y))

kmeans = KMeans(n_clusters=2, n_init='auto') # we are trying with K or n_clusters value as 2
kmeans.fit(data)

c=kmeans.labels_
print(c)
```

```python
plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

Evaluating the "Silhouette Coefficient" and selecting the value of n_clusters

```python
from sklearn import metrics
metrics.silhouette_score(data, c, metric='euclidean')
```



*  For n_cluster value 4 we are getting the Silhouette Coefficient as 0.41341316599442507

*   For n_cluster value 3 we are getting the Silhouette Coefficient as 0.38568424014359


*   For n_cluster value 2 we are getting the Silhouette Coefficient as 0.5417840060343528

we can observe that for n_cluster value 2 we are getting a good cluster as per Coefficient



















**We will try to find the clusters using high dimensional data**

```python
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
z = [15, 18, 22, 21, 21, 19, 15, 16, 16, 17]

from mpl_toolkits import mplot3d
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, color = "red")
plt.show()
```

Elbow method

```python
from sklearn.cluster import KMeans

data = list(zip(x, y, z))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

By the above graph we can observe that 2, 3 and 4 can be a good selection for the n_clusters value

Now we will add the K-Means algorithm with scikit learn library and visualize the clusters

```python
from sklearn.cluster import KMeans

data = list(zip(x, y, z))

kmeans = KMeans(n_clusters=4, n_init='auto') # Taking the n_clusters value as 4
kmeans.fit(data)

c=kmeans.labels_
print(c)
```

Visualizing the clusters in 3D

```python
ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z, c=kmeans.labels_)
plt.show()
```

Evaluating the Davies-Bouldin Index and identifying a n_clusters value

```python
metrics.davies_bouldin_score(data, c)
```

In this 3D dataset

*   If we take n_clusters value as 2 we get Davies-Bouldin Index as 0.8027808253784411
*    If we take n_clusters value as 3 we get Davies-Bouldin Index as 0.6281585750054066

*   If we take n_clusters value as 4 we get Davies-Bouldin Index as 0.6151120269650168

We can observe that taking n_clusters value as 4 we are getting the Davies-Bouldin Index closer to 0 which indicate a better partition.
